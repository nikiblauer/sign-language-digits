# Sign language digit classification
This notebook focuses on developing a deep learning model to classify digits in sign language using images. It leverages PyTorch for model building, training, and evaluation. The primary goal is to recognize hand gestures representing numbers (0-9)

## Imports
This section imports all necessary libraries and dependencies required for data handling, visualization, and model training. Key libraries include PyTorch for deep learning operations, NumPy for numerical computations, Matplotlib for data visualization, and additional utilities to streamline data preprocessing and model evaluation.

```python
import os
import sys


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import ImageGrid



import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, Dataset


from torchvision.utils import make_grid
from torchvision.io import read_image
from torchvision.transforms import transforms
```

## Prepare data for PyTorch
In this section, the dataset is prepared for use with PyTorch. This includes converting images and labels into PyTorch-compatible tensors, splitting the dataset into training and testing sets, and applying necessary transformations such as normalization. DataLoaders are created to efficiently manage batch processing during training and evaluation.

```python
class SignDataset(Dataset):
    def __init__(self, data_dir, train, transform=None, target_transform=None):
        super().__init__()
        
        self.data_dir = data_dir

        if train:
            self.labels = pd.read_csv(os.path.join(data_dir, 'train.csv'))
        else:
            self.labels = pd.read_csv(os.path.join(data_dir, 'test.csv'))

        self.transform = transform
        self.target_transform = target_transform


    def __len__(self):
        return len(self.labels)
        

    def __getitem__(self, index):
        image = read_image(os.path.join(self.data_dir, 'images', self.labels.iloc[index, 0])) / 255.0
        label = self.labels.iloc[index, 1]

        if self.transform:
            image = self.transform(image)
        if self.target_transform:
            label = self.target_transform(label)

        return image, label


transform = transforms.Compose([
    transforms.RandomRotation(degrees=10),      # Random rotation 5 degrees
    transforms.Resize((100, 100)),
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
])

train_dataset = SignDataset('data', train=True, transform=transform)
test_dataset = SignDataset('data', train=False, transform=transforms.Compose([transforms.Resize((100, 100)), transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])]))

batch_size = 5
train_dataloader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_dataloader = DataLoader(test_dataset, batch_size=batch_size)
        
```

### Augemented Training Images
To enhance the robustness of the model, data augmentation techniques are applied to the training images. Augmentation methods such as rotations, flips, and brightness adjustments are introduced to increase dataset variability, helping the model generalize better to unseen data.

```python

def imshow(img):
    img = img / 2 + 0.5     # unnormalize
    npimg = img.numpy()
    plt.imshow(np.transpose(npimg, (1, 2, 0)))
    plt.axis('off')  
    plt.show()


next(iter(train_dataloader))
images, labels = next(iter(train_dataloader))
imshow(make_grid(images[:4]))

classes = ('Zero', 'One', 'Two', 'Three', 'Four', 'Five',
           'Six', 'Seven', 'Eight', 'Nine')


print(' '.join(f'{classes[labels[j]]:20s}' for j in range(4)))
```

# Defining Neural Network
This section outlines the architecture of the neural network used for classification. The network consists of multiple layers, including convolutional layers for feature extraction, pooling layers for downsampling, and fully connected layers for final classification. The chosen architecture aims to balance complexity and performance for accurate digit recognition.

```python
class SignNetwork(nn.Module):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)

        self.conv1 = nn.Conv2d(3, 32, 3)
        self.conv2 = nn.Conv2d(32, 64, 3)
        self.conv3 = nn.Conv2d(64, 128, 3)
        self.pool = nn.MaxPool2d(2, 2)
        
        self.fc1 = nn.Linear(128 * 10 * 10, 256)
        self.fc2 = nn.Linear(256, 10)

    
    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = self.pool(F.relu(self.conv3(x)))
        x = x.view(x.size(0), -1)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)

        return x



model = SignNetwork()

model
```

## Train and Test model
The model is trained using the prepared data, with the training process involving optimization of the network's weights based on the loss function. Training metrics, such as loss and accuracy, are tracked and visualized to monitor progress. The model is then evaluated on the test set to assess its performance on unseen data.

```python
def train(dataloader, model, criterion, optimizer):
    model.train()
    total_batches = len(dataloader)
    for batch, (X, y) in enumerate(dataloader):
        outputs = model(X)
        loss = criterion(outputs, y)

        loss.backward()
        optimizer.step()
        optimizer.zero_grad()

        if batch % 5 == 0:
            sys.stdout.write(f"\r[{batch}/{total_batches}], Loss: {loss}")  # Overwrites the same line
            sys.stdout.flush()


def test(dataloader, model):
    model.eval()
    correct = 0
    with torch.no_grad():
        for X, y in dataloader:
            outputs = model(X)
            print(outputs.argmax(1))
            print(y)
            correct += (outputs.argmax(1) == y).sum().item()

    print(f"Accuracy: {correct / len(dataloader.dataset)}")
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.001, momentum=0.9)
epochs = 10


device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}\n")

for t in range(epochs):
    print(f"Epoch: {t+1} -------------------")
    train(train_dataloader, model=model, criterion=criterion, optimizer=optimizer)
    test(test_dataloader, model=model)
    print(f"\n\n")

```
